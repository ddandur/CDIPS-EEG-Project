```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import pandas
%matplotlib inline
```

```python
#load data
sample = scipy.io.loadmat('sample_clip.mat')
```

```python
#the (i,j) component of the correlation matrix is the correlation between electrode i with electrode j
ecog_rec = pandas.DataFrame(sample['data'])
correlation_matrix = ecog_rec.T.corr()
```

```python
#color plot of the 16 ecog electrodes correlations
plt.imshow(correlation_matrix,interpolation='nearest')
plt.title('Correlations of electrodes')
plt.xlabel('electrode number')
plt.ylabel('electrode number')
plt.colorbar()
plt.show()
```

```python
#this may or may not be useful, but if you look at the correlation in time you find this
plt.imshow(ecog_rec.corr())
plt.title('Correlations of time points')
plt.xlabel('time (2.5ms)')
plt.ylabel('time (2.5ms)')
plt.colorbar()
plt.show()
```

```python
electrode_statistics = pandas.DataFrame(data={'mean':ecog_rec.T.mean(),'variance': ecog_rec.T.var(), 'minimum': ecog_rec.T.min(),'maximum': ecog_rec.T.max(), 'standard deviation': np.sqrt(ecog_rec.T.var())})

electrode_statistics
```
